```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, silhouette_samples, davies_bouldin_score
from sklearn.decomposition import PCA


import warnings
warnings.filterwarnings('ignore')
sns.set_style('whitegrid')
```

# Data Preprocessing

```python
import pandas as pd
#STEP-2: loading dataset
data = pd.read_csv("Mall_Customers.xls") 
```

```python
data = data.rename(columns={'Annual Income (k$)': 'Annual Income', 'Spending Score (1-100)': 'Spending Score'})
data['Gender'] = data['Gender'].map({'Male': 1, 'Female': 0})
data = data.drop('CustomerID', axis=1)
```

```python
data
```

## Exploratory Data Analysis (EDA)

## Gender distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=data)
plt.title('Gender Distribution')
plt.xticks([0,1], ['Female', 'Male'])
plt.show()
```

## Pie chart for Gender

```python
values = data['Gender'].value_counts().sort_index()
labels = ['Female', 'Male']
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, explode=(0.05,0))
plt.title('Gender Distribution')
plt.show()
```

## Box Plot

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=data[['Age', 'Annual Income', 'Spending Score']])
plt.title("Boxplots of Age, Annual Income, Spending Score")
plt.show()
```

# Distributions and Pairplots

## Histograms with KDE 

```python
plt.figure(figsize=(16,4))
for i, col in enumerate(['Age', 'Annual Income', 'Spending Score']):
    plt.subplot(1, 3, i+1)
    sns.histplot(data[col], bins=15, kde=True)
    plt.title(f'{col} Distribution')
plt.tight_layout()
plt.show()
```

## PAIRPLOT

```python
sns.pairplot(data, vars=['Age','Annual Income','Spending Score'], hue='Gender', height=2.5)
plt.show()
```
